```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 60)
```

## Data importation

```python
#importing columns names
colums=[]
with open("columns.txt","r") as doc:
    lines=doc.readlines()
    for line in lines :
        colums.append(line.split(":")[0])
print(colums)
```

```python
df=pd.read_csv("data.csv",names=colums)
df
```

## Data exploration :

### correlation :

```python
sns.pairplot(df)
```

```python
corr=df.corr()
ax=sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns,
            cmap=sns.diverging_palette(220, 10, as_cmap=True))
ax.set(title="all variables correlation heatmap");
```

```python
#top correlation with class
k=10 #number of variables
cols=corr.nlargest(k,'class')['class'].index
cm=np.corrcoef(df[cols].values.T)
ax=sns.heatmap(cm,yticklabels=cols.values,
                xticklabels=cols.values,
                cmap=sns.diverging_palette(220, 10, as_cmap=True),
                annot=True)
ax.set(title=f"top {k} variables correlation with class heatmap");
plt.show()
```

The correlation between class and the top 10 variable is low

```python
#pair plot top 4 most correlated var
sns.pairplot(df[cols[:5].values]);
```
